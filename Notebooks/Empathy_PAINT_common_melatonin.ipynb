# load alighment

```
pos=pd.read_csv('/dbfs/mnt/database-datalakehouse/alignment/AlignedPeaks_rLC_pos_empathy_PAINT.csv')
neg=pd.read_csv('/dbfs/mnt/database-datalakehouse/alignment/AlignedPeaks_rLC_neg_empathy_PAINT.csv')
```

# match alignment

```
E_P_pos_list= ['rLC_pos_mtb_' + x for x in pos.loc[1:,'empathy'].astype(str)]
E_P_neg_list= ['rLC_neg_mtb_' + x for x in neg.loc[1:,'empathy'].astype(str)]

P_pos_list= ['rLC_pos_mtb_' + x for x in pos.loc[1:,'paint_rlc_pos_Most_likely'].astype(str)]
P_neg_list= ['rLC_neg_mtb_' + x for x in neg.loc[1:,'paint_rlc_neg_Most_likely'].astype(str)]

E_P_pos_list.extend(E_P_neg_list)
P_pos_list.extend(P_neg_list)

print(len(E_P_pos_list))
print(len(P_pos_list))
```

```
mdict = dict(zip(P_pos_list, E_P_pos_list))
```

```
aligned_emp_lmer=empathy_lmer_std[empathy_lmer_std.index.isin(E_P_pos_list)]
aligned_pai_lmer=piant_lmer_std[piant_lmer_std.index.isin(P_pos_list)]
```

```
aligned_pai_lmer['PAINT']=aligned_pai_lmer.index
aligned_pai_lmer['EMPATHY']=[mdict[p] for p in aligned_pai_lmer.index]
```

```
aligned_pai_lmer.columns='PAINT_'+aligned_pai_lmer.columns
aligned_pai_lmer.head(2)
```

```
aligned_emp_lmer.columns='EMPATHY_'+aligned_emp_lmer.columns
aligned_emp_lmer.head(2)
```

```
aligned_df=pd.merge(left=aligned_pai_lmer,right=aligned_emp_lmer,
                    left_on=aligned_pai_lmer['PAINT_EMPATHY'],
                    right_on=aligned_emp_lmer.index,
                    how='left'
                   )
print(aligned_df.shape)
aligned_df.head(2)
```

```
aligned_df.to_csv('/dbfs/mnt/client-002sap21p024-paint/04_data_analysis/empathy_paint_aligned_lmer.csv',index_label=False)
```

# select top marker

```
aligned_df=pd.read_csv('/dbfs/mnt/client-002sap21p024-paint/04_data_analysis/empathy_paint_aligned_lmer.csv')
```

```
Melatonin_df = aligned_df.loc[:,['EMPATHY_HGroupB','PAINT_HGroupB','EMPATHY_HGroupB_Pvalue','PAINT_HGroupB_Pvalue','PAINT_EMPATHY','PAINT_PAINT']].sort_values(['EMPATHY_HGroupB_Pvalue','PAINT_HGroupB_Pvalue'])
```

```
# p value < 0.05, same direction
low_p_Melatonin_df=Melatonin_df[(Melatonin_df.PAINT_HGroupB * Melatonin_df.EMPATHY_HGroupB >=0) &
                                (Melatonin_df.EMPATHY_HGroupB_Pvalue<0.05)&
                                (Melatonin_df.PAINT_HGroupB_Pvalue<0.05)
                               ]
low_p_Melatonin_df.shape
```

# attach MZ, RT, High sample abundance

```
PAINT_MZRT=Paint_df.copy()
PAINT_MZRT['PAINT_Median_Raw_abundance']=Paint_df.iloc[:,3:].median(axis=1)
PAINT_MZRT['PAINT_Max_Raw_abundance']=Paint_df.iloc[:,3:].max(axis=1)
PAINT_MZRT['PAINT_Max_Raw_abundance_sample']=Paint_df.iloc[:,3:].idxmax(axis=1)
PAINT_MZRT['PAINT_MZ']=Paint_df['MZ']
PAINT_MZRT['PAINT_RT']=Paint_df['RT']
```

```
EMPATHY_MZRT=emp_df.copy()
EMPATHY_MZRT['EMPATHY_Median_Raw_abundance']=emp_df.iloc[:,3:].median(axis=1)
EMPATHY_MZRT['EMPATHY_Max_Raw_abundance']=emp_df.iloc[:,3:].max(axis=1)
EMPATHY_MZRT['EMPATHY_Max_Raw_abundance_sample']=emp_df.iloc[:,3:].idxmax(axis=1)
EMPATHY_MZRT['EMPATHY_MZ']=emp_df['MZ']
EMPATHY_MZRT['EMPATHY_RT']=emp_df['RT']
```

```
MZinfo_low_p_Melatonin_df=pd.merge(left=low_p_Melatonin_df,right=EMPATHY_MZRT.loc[:,['EMPATHY_MZ','EMPATHY_RT','EMPATHY_Median_Raw_abundance','EMPATHY_Max_Raw_abundance','EMPATHY_Max_Raw_abundance_sample']],
                                   left_on=low_p_Melatonin_df.PAINT_EMPATHY,
                                   right_on=emp_df.Metabolite_Feature_Label,
                                   how='left')
MZinfo_low_p_Melatonin_df.drop(['key_0'],axis=1,inplace=True)
```

```
MZinfo_low_p_Melatonin_df=pd.merge(left=MZinfo_low_p_Melatonin_df,right=PAINT_MZRT.loc[:,['PAINT_MZ','PAINT_RT','PAINT_Median_Raw_abundance','PAINT_Max_Raw_abundance','PAINT_Max_Raw_abundance_sample']],
                                   left_on=low_p_Melatonin_df.PAINT_PAINT,
                                   right_on=Paint_df.Metabolite_Feature_Label,
                                   how='left')
MZinfo_low_p_Melatonin_df.drop(['key_0'],axis=1,inplace=True)
```

# filter by 20 ppm

```
MZinfo_low_p_Melatonin_df=pd.read_csv('/dbfs/mnt/client-002sap21p024-paint/04_data_analysis/empathy_paint_aligned_lmer_metatonin_res_MZinfo.csv')
```

```
MZinfo_low_p_Melatonin_df['ppm20_filter']='Yes'
```

```
for m in list(MZinfo_low_p_Melatonin_df.PAINT_PAINT):
    paint_mz=MZinfo_low_p_Melatonin_df.loc[MZinfo_low_p_Melatonin_df.PAINT_PAINT==m,'PAINT_MZ']
    emp_mz=MZinfo_low_p_Melatonin_df.loc[MZinfo_low_p_Melatonin_df.PAINT_PAINT==m,'EMPATHY_MZ']
    error=abs(paint_mz-emp_mz)
    ppm20=20*emp_mz/1e6
    if (error.values-ppm20.values)>=0:
#         print(m)
        MZinfo_low_p_Melatonin_df.loc[MZinfo_low_p_Melatonin_df.PAINT_PAINT==m,'ppm20_filter']='No'
```

```
MZinfo_low_p_Melatonin_df.ppm20_filter.value_counts()
```

```
filtered_MZinfo_low_p_Melatonin_df=MZinfo_low_p_Melatonin_df[MZinfo_low_p_Melatonin_df.ppm20_filter=='Yes']
```
